```python
В выбираемый датасет добавляем атрибут: Customers - кол-во покупателей. 
Добавление происходит через функцию =СЛЧИС()*100 - случайное число больше или равное 0, но меньше 100.
Теперь напишем на python программу:

В ходе анализа данных мы добавили атрибут - кол-во покупателей. Далее мы очистили данные от выбросов (таких значений оказалось 0), и проделали
визуализацию данных, а после применили метод KMeans. Это поможет более точно определить спрос на разного вида жилье.
```

```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Загрузка данных с правильным разделителем
file_path = r"C:\Users\defau\OneDrive\Desktop\house price prediction dataset.csv"
data = pd.read_csv(file_path)

# Проверка названий столбцов
print("Названия столбцов:\n", data.columns)

# Удаление лишних пробелов из названий столбцов
data.columns = data.columns.str.strip()

# Проверка названий столбцов после очистки
print("Названия столбцов после очистки:\n", data.columns)

# Причесывание датасета
## Удаление дубликатов
data.drop_duplicates(inplace=True)

## Обработка пропущенных значений
# Проверка на пропущенные значения
missing_values = data.isnull().sum()
print("Пропущенные значения:\n", missing_values)

# Удаление строк с пропущенными значениями
data.dropna(inplace=True)

## Удаление выбросов
# Выбор только числовых столбцов для вычисления квартилей
numeric_data = data.select_dtypes(include=[np.number])

Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1

# Удаление выбросов
data = data[~((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)]

## Поиск кривых данных
# Визуализация для поиска аномалий
plt.figure(figsize=(10, 6))
sns.boxplot(data['Price'])  # Используйте правильное имя столбца
plt.title('Boxplot цен')
plt.show()

# Построение графиков зависимости
## Пример: зависимость цены от площади
plt.figure(figsize=(10, 6))
plt.scatter(data['Area'], data['Price'])  # Используйте правильные имена столбцов
plt.title('Зависимость цены от площади')
plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.show()

# Применение KMeans
## Выбор признаков для кластеризации (добавлен столбец Customers)
features = data[['Area', 'Price', 'Customers']]  # Используйте правильные имена столбцов

## Стандартизация данных
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

## Определение оптимального числа кластеров с помощью метода локтя
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()

## Применение KMeans
optimal_clusters = 3  # Замените на оптимальное количество кластеров
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['cluster'] = kmeans.fit_predict(scaled_features)

## Визуализация кластеров
plt.figure(figsize=(10, 6))
plt.scatter(data['Area'], data['Price'], c=data['cluster'], cmap='viridis')
plt.title('Кластеры домов')
plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.colorbar(label='Кластер')
plt.show()

# Дополнительно: визуализация зависимости количества покупателей от цены и площади
plt.figure(figsize=(10, 6))
plt.scatter(data['Area'], data['Customers'], c=data['cluster'], cmap='viridis')
plt.title('Количество покупателей по площади')
plt.xlabel('Площадь')
plt.ylabel('Количество покупателей')
plt.colorbar(label='Кластер')
plt.show()
```
